```python
# Split Input and Output

# How to spread the Linear ploting

# One coloum viewed or Independent viewed or get input values

indep = dataset[["YearsExperience"]];

#print(indep);

indep

```

```python
# Dependent values or Output viewed


dep = dataset[["Salary"]];

#print(dep);

dep
```

```python
# How to draw a graph in Python or Scatter

import matplotlib.pyplot as plt;

plt.scatter(indep,dep);

# x-axis and y-axis showing

plt.xlabel("Years of Experience");

plt.ylabel("Salary");

plt.show(); #not necessary
```

```python
# Split train and set

from sklearn.model_selection import train_test_split;

x_train, x_test, y_train, y_set = train_test_split(indep, dep, test_size = 0.30, random_state = 0);

#print(x_train);

x_train
```

```python
x_train, x_test, y_train, y_set = train_test_split(indep, dep, test_size = 0.30, random_state = 0);

print(x_train); # Random state value adjust 
```

```python
# Model Creation

from sklearn.linear_model import LinearRegression;

regressor = LinearRegression();

regressor.fit(x_train, y_train); 

weight = regressor.coef_;   # weight calculate

print(weight); 
```

```python
# Bias calculate or origin value or starting value or Fresher salary


bias = regressor.intercept_;   #bias calculate

print(bias);
```

```python
# Test set or Predict test set 

y_pred = regressor.predict(x_test);

print(y_pred);
```

```python
# Creating new Data Set or create table for Actual Value

predActual = pd.DataFrame(index=range(0,8))

predActual["Actual Value"] = y_set;

y_set.index=range(0,9); # Index Alter in Order wise / not nessesary if you have already ordered

#print(y_set);

y_set
```

```python
predActual
```

```python
# create table for Predict Value

y_pred_table = pd.DataFrame(y_pred, columns = ["Predict Value"]);

y_pred_table
```

```python

predActual["Predict Value"] = y_pred_table;

predActual
```

```python
# Evaluation metric or r2_score

from sklearn.metrics import r2_score;

r = r2_score(y_set, y_pred);

r    # if r value nearly to 1 is Good Model / nearly 0 - Poor Model


```

```python
# Predicted the best model

print(r);
```
